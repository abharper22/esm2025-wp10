```python
def average_dict(input):
    output=sum( np.array(input[clim] )[0][:] for clim in ESMS) / 3
    return(output)
```

```python
REGIONS=['Global','Tropics','NH-Temperate']
MODELS=['jules','clm','jsbach']
LANDUSES=['noluc','agtobio','nattobio']
SSPS=['ssp126','ssp370']
ESMS=['mpi-esm1-2-hr','ukesm1-0-ll','ipsl-cm6a-lr']
VARIABLES=['fHarvest']
# JULES      fBioHarvestpft - extract PFT 13
# JSBACH     cHbpharv
# CLM        fHarvest: loss from 1-yr crop product pool, kgC/m2/s

years = np.arange(2015,2101)     
nyears = len(years)

# Option: 
# plot_total means plot the cumulative total harvests in PgC. Problem with this is 
#            that it will always show higher totals for nat to bio since bioenergy
#            crops are placed in all natural land.
# plot_total=False means annual harvests are plotted in tC/ha.

plot_total = False


#Set up plotting stuff
#Each land model needs a different color
fig,axes=plt.subplots(1,1, figsize=[4,3])
fig.suptitle('Global: No land use change',fontsize=12)
ix=0

LU_SCENARIO_MEAN = {var: {lsm_mod: {region: {landuse: {ssp_scen: [] for ssp_scen in SSPS} 
                    for landuse in LANDUSES} for region in REGIONS} for lsm_mod in MODELS} 
                    for var in VARIABLES}
landuse_exists = {lsm_mod: {landuse: [True] for landuse in LANDUSES} for lsm_mod in MODELS}

variable_exists = {var: {lsm_mod: [True] for lsm_mod in MODELS} for var in VARIABLES}


test_mean={ssp: [] for ssp in SSPS}
colors = {model: [] for model in MODELS}
colors['jules'] = '#F97306'
colors['clm'] = '#C5C9C7'
colors['jsbach'] = '#D1B26F'
colors['orchidee'] = '#A9561E'
#https://matplotlib.org/stable/users/explain/colors/colors.html

linestyles = {ssp: [] for ssp in SSPS}
linestyles['ssp126'] = '-'
linestyles['ssp370'] = '--'

#cVeg and cSoil are easy
for var in VARIABLES:

    # set axes
    ax=axes
    if plot_total:
        ax.set_ylabel('Cumulative harvests (PgC) ',fontsize=10)
    else:
        ax.set_ylabel('Average harvests (tC/ha) ',fontsize=10)
    ax.set_xlim([2015,2100])
    ax.grid(True) 
    
    for model in MODELS:
        #First get no land use change (noluc)
        if variable_exists[var][model]:
            if model=='jules':
                varname='fBioHarvestpft'
                units='tC/ha' #originally kgC/m2/s
                scenario='nolucbio'
            elif model=='jsbach':
                varname='cHbpharv'
                units='tC/ha'  #originally kgC/m2
                scenario='noluc'
            else:
                varname=var
                units='tC/ha'  #originally kgC/m2/s (CLM)
                scenario='noluc'

            noluc,noluc_tr,noluc_te = retrieve_data_generic(varname,units,scenario,model)
            LU_SCENARIO_MEAN[var][model]['Global']['noluc']['ssp126']=average_dict(noluc['ssp126'])
            LU_SCENARIO_MEAN[var][model]['Global']['noluc']['ssp370']=average_dict(noluc['ssp370'])
            LU_SCENARIO_MEAN[var][model]['Tropics']['noluc']['ssp126']=average_dict(noluc_tr['ssp126'])
            LU_SCENARIO_MEAN[var][model]['Tropics']['noluc']['ssp370']=average_dict(noluc_tr['ssp370'])
            LU_SCENARIO_MEAN[var][model]['NH-Temperate']['noluc']['ssp126']=average_dict(noluc_te['ssp126'])
            LU_SCENARIO_MEAN[var][model]['NH-Temperate']['noluc']['ssp370']=average_dict(noluc_te['ssp370'])
 
            #Next get land use change scenarios
            for landuse in LANDUSES[1:]:
                if (landuse_exists[model][landuse]):
                    if model=='jules':
                        varname='fBioHarvestpft'
                        units='tC/ha'
                    elif model=='jsbach':
                        varname='cHbpharv'
                        units='tC/ha'
                    else:
                        varname=var
                        units='tC/ha'

                    cdata,cdata_tr,cdata_te = retrieve_data_generic(varname,units,landuse,model)
                    LU_SCENARIO_MEAN[var][model]['Global'][landuse]['ssp126']=average_dict(cdata['ssp126'])
                    LU_SCENARIO_MEAN[var][model]['Global'][landuse]['ssp370']=average_dict(cdata['ssp370'])
                    LU_SCENARIO_MEAN[var][model]['Tropics'][landuse]['ssp126']=average_dict(cdata_tr['ssp126'])
                    LU_SCENARIO_MEAN[var][model]['Tropics'][landuse]['ssp370']=average_dict(cdata_tr['ssp370'])
                    LU_SCENARIO_MEAN[var][model]['NH-Temperate'][landuse]['ssp126']=average_dict(cdata_te['ssp126'])
                    LU_SCENARIO_MEAN[var][model]['NH-Temperate'][landuse]['ssp370']=average_dict(cdata_te['ssp370'])
                        
            # FIRST PLOT THE NO LAND USE CHANGE SCENARIO.
            for ssp in SSPS:
                if model=='jsbach':
                    endyear=nyears-1
                else:
                    endyear=nyears
                print(endyear)
                if plot_total==True:
                    if model=='jsbach':
                        totharvest = LU_SCENARIO_MEAN[var][model]['Global']['noluc'][ssp]
                    else:
                        totharvest = np.cumsum(LU_SCENARIO_MEAN[var][model]['Global']['noluc'][ssp])
                else:
                    if model=='jsbach':
                        for j in range(0,endyear-1):
                            totharvest[j]=LU_SCENARIO_MEAN[var][model]['Global']['noluc'][ssp][j+1]- \
                                          LU_SCENARIO_MEAN[var][model]['Global']['noluc'][ssp][j]
                            #print(j,years[j],totharvest[j])
                        totharvest[endyear-1]=totharvest[endyear-2]
                        #print(endyear,years[endyear-1],totharvest[endyear-1])
                    else:
                        totharvest = LU_SCENARIO_MEAN[var][model]['Global']['noluc'][ssp]
                        
                ax.plot(years[:endyear], totharvest[:endyear]-totharvest[0], label=model+' '+ssp,
                            c=colors[model],ls=linestyles[ssp],lw=2)

    ix+=1


handles,labels = ax.get_legend_handles_labels()
fig.legend(handles,labels,ncol=2,loc='upper left',bbox_to_anchor=(0.55, 0.0),fontsize=10)
fig.tight_layout()

plt.savefig('/bdd/ESM2025/WP10/plots/fharvest_noluc_Global_tC.png', format="png")
```

```python
# PLOT THE LANDUSE SCENARIO DIFFERENCE FROM THE NOLUC SCENARIO

landuse_title=['noluc',' Agriculture to bioenergy', ' Natural to bioenergy']

plotdiff = True

i1=1
for landuse in LANDUSES[1:]:
    print(landuse)     
    #Set up plotting stuff
    #Each land use scenario has a new figure
    fig2,axes2=plt.subplots(1,3, figsize=[12,3])
    
    ix=0    
    for region in REGIONS:

        for var in VARIABLES:

            # set axes
            ax=axes2[ix]
            if plotdiff:
                if plot_total:
                    ax.set_ylabel('Cumulative harvests (Bio-noluc) (PgC)',fontsize=10)
                else:
                    ax.set_ylabel('Average harvests (tC/ha) ',fontsize=10)                  
            else:
                if plot_total:
                    ax.set_ylabel('Cumulative harvests (PgC) ',fontsize=10)
                else:
                    ax.set_ylabel('Average harvests (tC/ha) ',fontsize=10)  
            ax.set_xlim([2015,2100])
            ax.grid(True) 
        
            for model in MODELS:
                if model=='jsbach':
                    endyear=nyears-1
                    totharvest = np.empty(nyears-1)
                    totharvest_noluc = np.empty(nyears-1)

                else:
                    endyear=nyears
                    totharvest = np.empty(nyears)
                    totharvest_noluc = np.empty(nyears)

                
                if (landuse_exists[model][landuse] and variable_exists[var][model]):
                    print(var,model,region,landuse)
                    for ssp in SSPS:
                        
                        if model=='jsbach':
                            if plot_total:
                                totharvest = LU_SCENARIO_MEAN[var][model][region][landuse][ssp]
                                totharvest_noluc = LU_SCENARIO_MEAN[var][model][region]['noluc'][ssp]
                            else:
                                for j in range(0,endyear-1):
                                    totharvest[j]=LU_SCENARIO_MEAN[var][model][region][landuse][ssp][j+1]- \
                                                  LU_SCENARIO_MEAN[var][model][region][landuse][ssp][j]
                                    totharvest_noluc[j]=LU_SCENARIO_MEAN[var][model][region]['noluc'][ssp][j+1]- \
                                                  LU_SCENARIO_MEAN[var][model][region]['noluc'][ssp][j]
                                totharvest[endyear-1]=totharvest[endyear-2]                                
                                totharvest_noluc[endyear-1]=totharvest_noluc[endyear-2]                                
                        else:
                            if plot_total:
                                totharvest = np.cumsum(LU_SCENARIO_MEAN[var][model][region][landuse][ssp])
                                totharvest_noluc = np.cumsum(LU_SCENARIO_MEAN[var][model][region]['noluc'][ssp])
                            else:
                                totharvest = LU_SCENARIO_MEAN[var][model][region][landuse][ssp]
                                totharvest_noluc = LU_SCENARIO_MEAN[var][model][region]['noluc'][ssp]
                                           
                        if (plotdiff):
                            plot_luc=(totharvest)-(totharvest_noluc)
                        else:
                            plot_luc=totharvest
                            
                        ax.plot(years[:endyear], plot_luc[:endyear], label=model+' '+ssp,
                                    c=colors[model],ls=linestyles[ssp],lw=2)
                        ax.set_title(region+landuse_title[i1],fontsize=10)
                        ax.set_ylim(0,20)
                        print(var,model,ssp,'2030s:',np.mean(totharvest[15:25]),
                              '2050s:',np.mean(totharvest[35:45]),
                              '2090s:',np.mean(totharvest[75:]))

                        #m,b = np.polyfit(years[endyear-20:endyear-1],plot_noluc[-20:-1], 1)
                        #print(var,model,ssp,'slope over last 20 years:',m)
        ix+=1



    handles2,labels2 = ax.get_legend_handles_labels()
    fig2.legend(handles2,labels2,ncol=2,loc='upper left',bbox_to_anchor=(0.38, 0.05),fontsize=10)
    fig2.tight_layout()

    if plotdiff:
        plt.savefig('/bdd/ESM2025/WP10/plots/fharvest_'+landuse+'-noluc_tC.png', format="png")
    else:
        plt.savefig('/bdd/ESM2025/WP10/plots/fharvest_'+landuse+'_tC.png', format="png")
    i1+=1
```
